```python
import json

input_file_path = "/home/hb/fine-tuning-alpaca/evaluation/eval_fill_the_blank.txt"
output_file_path = "bgp_test_fill_the_blank.json"

questions = []
current_question = {}
is_question = False

with open(input_file_path, "r") as input_file:
    for line in input_file:
        line = line.strip()
        if line.startswith("Answer: "):
            current_question["answer"] = line[len("Answer: "):]
            questions.append(current_question)
            current_question = {}
            is_question = False
        elif line:
            if is_question:
                current_question["question"] += " " + line
            else:
                current_question["question"] = line
                is_question = True

# Write the questions to a JSON file
with open(output_file_path, "w") as output_file:
    json.dump(questions, output_file, indent=4)

print("JSON file created successfully.")
```

```python
import json

input_file_path = "/home/hb/fine-tuning-alpaca/evaluation/eval_mult_choice.txt"
output_file_path = "bgp_test_multiple_choice.json"

questions = []

with open(input_file_path, "r") as input_file:
    lines = input_file.read().strip().split("\n\n")
    
    for q_text in lines:
        q_lines = q_text.strip().split("\n")
        question = q_lines[0]
        options = q_lines[1:-1]
        answer = q_lines[-1].replace("Answer: ", "")
        
        question_dict = {
            "question": question,
            "options": options,
            "answer": answer
        }
        questions.append(question_dict)

# Write the questions to a JSON file
with open(output_file_path, "w") as output_file:
    json.dump(questions, output_file, indent=4)

print("JSON file created successfully.")
```
